```python
# Import libraries
import pandas as pd 
import numpy as np 
import matplotlib.pylab as plt 
import seaborn as sns 

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python

a = np.linspace(10,200,500)
b = np.random.random_sample(500)

print(len(a))
len(b)
```

```python
#basic scatter plot
```

```python
plt.scatter(a,b)
```

```python
plt.rcParams['figure.figsize']= (10,5)
plt.rcParams['figure.dpi'] = 300
```

```python
plt.scatter(a,b, color = 'r', edgecolor = 'blue')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('A vs B ')
```

```python
#read data
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
#set style and theme for scatter plot
sns.set_theme(style='darkgrid')
```

```python
sns.scatterplot(x = df['price'],y = df['engine-size'], color = 'c', edgecolor = 'black', hue= df['body-style'])
```

```python
sns.scatterplot(x = df['price'],y = df['engine-size'], color = 'c', edgecolor = 'black', hue= df['engine-location'])
```

```python
sns.scatterplot(x = df['price'],y = df['engine-size'], color = 'c', edgecolor = 'black', hue= df['body-style'])
plt.xlabel('Price')
plt.ylabel('Engine-size')
plt.title('Price VS Engine-size',size=25)
```

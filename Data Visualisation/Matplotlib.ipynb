#
- Xlim : is used to define limit for x variable
- Ylim : is used to define limit for y variable 

```python
area = [1000,1520,2000,3000,5210]
price = [10,15,12,25,30]

plt.plot(area,price)
plt.grid()

plt.xlim(1000,6000)
plt.ylim(5,50)

plt.xlabel( "Area of the House")
plt.ylabel("Price of the House")
```

#
- Xticks =  rotate lable
- Yticks =  rotate lable

```python
area = [1000,1520,2000,3000,5210]
price = [10,15,12,25,30]

plt.plot(area,price)

plt.xticks(rotation = 45)
plt.yticks(rotation = 10)



plt.xlabel( "Area of the House")
plt.ylabel("Price of the House")
```

# 
- Multiple plots : multiple line in one chart called multiple plots.
- Lrgends : gives lables to plot.
- Colours : used to give colour to plot.

```python
x  = np.arange(1,20)
y1 = np.sqrt(x)
y2 = np.log(x)

plt.plot(x,y1, color = 'b', label = 'Square root')
plt.plot(x,y2, color = 'g', label = 'Log')
plt.legend
```

# Matplolib  Histogram
- A histogrsm is a graph showing frequency disrtibution.
- It is the graph showing the number of observation within each given enterval.

```python
import pandas as pd 
```

```python
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('Automobile_data.csv')
df.head()
```

```python
df.shape[1]
```

```python
df.describe().T
#T is used for chaning row into coloumn or coloumn into row.
```

```python

df.isna().sum()
```

```python
plt.hist(df['width'])
plt.title('Width Frequency Table')
plt.xlabel('Width')
plt.ylabel('frequency')
```

```python
width_mean = df['width'].mean()
width_median = df['width'].median()

df['width'].hist(bins=10)
plt.axvline(width_mean, color = 'y')
plt.axvline(width_median, color = 'r')
plt.ylabel('Frequency')
plt.xlabel('Width')
```
